```python
# Week 1 Bonus: Polynomial Regression to fit Exponential Dataset

import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

# Exponential dataset
X_exp = np.array([[1], [2], [3], [4], [5], [6]])
y_exp = np.array([100, 150, 225, 337, 506, 759])

# Transform features to polynomial (degree=2)
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X_exp)

# Train Linear Regression on polynomial features
model_poly = LinearRegression()
model_poly.fit(X_poly, y_exp)

# Predict for a new value
X_new_exp = np.array([[7]])
X_new_poly = poly.transform(X_new_exp)
y_pred_poly = model_poly.predict(X_new_poly)
print(f"Polynomial Regression Prediction for 7: {y_pred_poly[0]:.2f}")

# Predict across a range for smooth curve
X_range = np.linspace(1, 7, 100).reshape(-1, 1)
X_range_poly = poly.transform(X_range)
y_range_pred = model_poly.predict(X_range_poly)

# Plot results
plt.scatter(X_exp, y_exp, color='blue', label='Actual Prices')
plt.plot(X_range, y_range_pred, color='red', label='Polynomial Regression Fit')
plt.scatter(X_new_exp, y_pred_poly, color='green', label='Prediction', s=100)
plt.title("Polynomial Regression - Exponential Dataset")
plt.xlabel("Size (1000 sq ft)")
plt.ylabel("Price ($1000s)")
plt.legend()
plt.grid(True)
plt.show()
```
