```python
# Week 1 Mini Project: Linear vs Exponential Dataset with Linear Regression

import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# ----------------------------------------
# 1. Linear Dataset
# ----------------------------------------
X_linear = np.array([[1], [2], [3], [4], [5], [6]])
y_linear = np.array([100, 120, 140, 160, 180, 200])

# Train Linear Regression
model_linear = LinearRegression()
model_linear.fit(X_linear, y_linear)

# Predict
X_new = np.array([[7]])
y_pred_linear = model_linear.predict(X_new)
print(f"Linear Dataset Prediction for 7: {y_pred_linear[0]:.2f}")

# Plot Linear Dataset
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.scatter(X_linear, y_linear, color='blue', label='Actual Prices')
plt.plot(X_linear, model_linear.predict(X_linear), color='red', label='Regression Line')
plt.scatter(X_new, y_pred_linear, color='green', label='Prediction', s=100)
plt.title("Linear Dataset - House Price Prediction")
plt.xlabel("Size (1000 sq ft)")
plt.ylabel("Price ($1000s)")
plt.legend()
plt.grid(True)

# ----------------------------------------
# 2. Exponential / Non-linear Dataset
# ----------------------------------------
X_exp = np.array([[1], [2], [3], [4], [5], [6]])
y_exp = np.array([100, 150, 225, 337, 506, 759])  # Exponential growth

# Train Linear Regression
model_exp = LinearRegression()
model_exp.fit(X_exp, y_exp)

# Predict
X_new_exp = np.array([[7]])
y_pred_exp = model_exp.predict(X_new_exp)
print(f"Exponential Dataset Prediction for 7: {y_pred_exp[0]:.2f}")

# Plot Exponential Dataset
plt.subplot(1,2,2)
plt.scatter(X_exp, y_exp, color='blue', label='Actual Prices')
plt.plot(X_exp, model_exp.predict(X_exp), color='red', label='Regression Line')
plt.scatter(X_new_exp, y_pred_exp, color='green', label='Prediction', s=100)
plt.title("Exponential Dataset - Linear Regression Fit")
plt.xlabel("Size (1000 sq ft)")
plt.ylabel("Price ($1000s)")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
